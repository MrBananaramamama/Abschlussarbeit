# Auswertung Anteil Aktienkapital

## Für den Index UBS 100 

```python
import pandas as pd
import matplotlib
```

```python
Firmen_Liste = pd.read_csv('Master_File_FERTIG.csv')  
Firmen_ListeX = pd.read_csv('Master_File_FERTIG.csv')  
```

```python
#Wie viel Prozent des Aktienkapitals/Stimmrechte ist bekannt?
Firmen_Liste['Anteil'] = Firmen_Liste['Anteil'].str.strip('%')
Firmen_Liste['Anteil'] = pd.to_numeric(Firmen_Liste['Anteil'], downcast='float')
#Es sind also 85 Firmen. Für den bekannten Besitz Teile Summe 'Anteil' durch 8500 Prozent.
Firmen_Liste['Anteil'].sum()/8500*100
```

```python
print('Von den 85 ausgewerteten Firmen sind ' +str(Firmen_Liste['Anteil'].sum()/8500*100) +' Prozent des Aktienkapitals bekannt.')
print('Somit befinden sich ' +str(100-Firmen_Liste['Anteil'].sum()/8500*100) +' Prozent der Aktien im Streubesitz.')
```

```python
#Welchen Ländern gehört wie viel? TOTAL AKTIENKAPITAL
NewDF = pd.DataFrame(Firmen_Liste.groupby('Country')['Anteil'].sum()/8500*100)
pd.DataFrame.sort_values(NewDF, by='Anteil', ascending=False)
```

```python
#Bekannt sind also gut 38,9 Prozent des gesamten Aktienkapitals, was 2546% Total entspricht. Weil, siehe unten. 
8500*0.389
```

```python
#Welchen Ländern gehört wie viel? Anteil am bekannten Aktienkapital.
NewDF = pd.DataFrame(Firmen_Liste.groupby('Country')['Anteil'].sum()/3306.5*100)
pd.DataFrame.sort_values(NewDF, by='Anteil', ascending=False)
```

```python
#Wem gehört am meisten? Anteil am bekannten Aktienkapital.

DF = pd.DataFrame(Firmen_Liste.groupby('BO')['Anteil'].sum()/3306.5*100)
pd.DataFrame.sort_values(DF, by='Anteil', ascending=False)
```

## Relative Auswertung SMI Expanded, mit Hilfe der Marktkapitalisierung der Firmen

```python
# Erst wenn man weiss, welchen Wert die jeweiligen Beteiligungen haben, kann man einevverlässliche Aussage zum Besitz treffen:
########## HIER FÜR SMI EXPANDED #############
#HIER GLEICHE ICH DIE WERTE MIT DER BEKANNTEN MARKTKAPITALISIERUNG DER FIRMEN AB. SO ENTSTEHT ERST EIN AUSSAGEKRÄFTIGES BILD.
#ES GILT ZU BEMERKEN, DASS DIE SMI FIRMEN 80 PROZENT DES GESAMTEN SIX AKTIENINDEXES IN SICH VEREINEN.
```

```python
SMI_MC_E = pd.read_csv('SMI Expanded mit Marktkapitalisierung.csv')
SMI_E = pd.read_csv('SMI Expanded Ganz.csv')
```

```python
SMI_Merge2 = pd.merge(SMI_MC_E, SMI_E, on='Company', how='inner')
```

```python
SMI_Merge2['Anteil in Mio. CHF'] = SMI_Merge2['Marktkapitalisierung, in Mio.'] /100* SMI_Merge2['Anteil']
```

```python
SMI_Merge2 = SMI_Merge2[['Company','Marktkapitalisierung, in Mio.', 'Anteil in Mio. CHF', 'BO', 'DS', 'Country']]
```

```python
SMI_Merge2['BO'] = SMI_Merge2['BO'].replace('BlackRock, Inc., New York, NY, U.S.A. (Mother Company)', 'Blackrock')
```

```python

#Wie viele Milliarden Besitz sind bekannt?
SMI_Merge2['Anteil in Mio. CHF'].sum()/1000
```

```python
#WELCHE LÄNDER BESITZEN AM MEISTEN AM SMI? 
SMI_Anteil_Mia2 = pd.DataFrame(SMI_Merge2.groupby('Country')['Anteil in Mio. CHF'].sum())
pd.DataFrame.sort_values(SMI_Anteil_Mia2, by='Anteil in Mio. CHF', ascending=False)
```

```python
SMI_Anteil_Mia_Personen2 = pd.DataFrame(SMI_Merge2.groupby('BO')['Anteil in Mio. CHF'].sum(normalize=True))
table = pd.DataFrame.sort_values(SMI_Anteil_Mia_Personen2, by='Anteil in Mio. CHF', ascending=False)
table.head(50)
```

```python
SMI_Anteil_Mia2 = pd.DataFrame(SMI_Merge2.groupby('Country')['Anteil in Mio. CHF'].sum()).plot(kind='pie', subplots=True, autopct='%1.1f%%', figsize=(10, 100))
```

```python
SMI_Anteil_Miax = pd.DataFrame(SMI_Merge2.groupby('BO')['Anteil in Mio. CHF'].sum())
```

```python
#ANTEIL PERSONEN IN PROZENT
new = SMI_Anteil_Miax['Anteil in Mio. CHF'].sum() / SMI_Anteil_Miax['Anteil in Mio. CHF']
```
